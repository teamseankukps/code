```python
import os
import numpy as np

def write_results_to_file(results, results_test, filename="yolo_results.txt"):
    with open(filename, "w") as f:
        f.write("=== Training Results ===\n")
        if hasattr(results, 'epoch'):
            f.write(f"Total epochs: {results.epoch}\n")
        if hasattr(results, 'best_epoch'):
            f.write(f"Best epoch: {results.best_epoch}\n")
        if hasattr(results, 'total_time'):
            f.write(f"Total training time: {results.total_time}\n")
        if hasattr(results, 'metrics'):
            for key, value in results.metrics.items():
                f.write(f"{key}: {value}\n")

        f.write("\n=== Validation Results ===\n")
        if hasattr(results_test, 'box'):
            f.write(f"mAP50: {results_test.box.map50:.4f}\n")
            f.write(f"mAP50-95: {results_test.box.map:.4f}\n")

            # Handle cases where values might be numpy arrays
            p = results_test.box.p
            r = results_test.box.r
            f1 = results_test.box.f1

            if isinstance(p, np.ndarray):
                f.write(f"Precision: {np.mean(p):.4f}\n")
            else:
                f.write(f"Precision: {p:.4f}\n")

            if isinstance(r, np.ndarray):
                f.write(f"Recall: {np.mean(r):.4f}\n")
            else:
                f.write(f"Recall: {r:.4f}\n")

            if isinstance(f1, np.ndarray):
                f.write(f"F1-score: {np.mean(f1):.4f}\n")
            else:
                f.write(f"F1-score: {f1:.4f}\n")

        if hasattr(results_test, 'nc'):
            f.write(f"Number of classes: {results_test.nc}\n")
        if hasattr(results_test, 'names'):
            f.write(f"Names of classes: {results_test.names}\n")
        if hasattr(results_test, 'speed'):
            f.write(f"Inference speed: {results_test.speed['inference']:.4f} ms\n")
            f.write(f"Preprocess speed: {results_test.speed['preprocess']:.4f} ms\n")
            f.write(f"Postprocess speed: {results_test.speed['postprocess']:.4f} ms\n")
        if hasattr(results_test, 'n'):
            f.write(f"Number of images: {results_test.n}\n")
        if hasattr(results_test, 'shape'):
            f.write(f"Image shape: {results_test.shape}\n")
        if hasattr(results_test, 'device'):
            f.write(f"Device used: {results_test.device}\n")
        if hasattr(results_test, 'data'):
            f.write(f"Data YAML path: {results_test.data}\n")
        if hasattr(results_test, 'task'):
            f.write(f"Task: {results_test.task}\n")
        if hasattr(results_test, 'batch_size'):
            f.write(f"Batch size: {results_test.batch_size}\n")

    print(f"Results have been written to {os.path.abspath(filename)}")

# ใช้ฟังก์ชันนี้หลังจากที่คุณได้ train และ validate โมเดลแล้ว
write_results_to_file(results, results_test)
```

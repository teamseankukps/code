```python
import os
from collections import defaultdict
```

```python
def count_yolo_labels(directory):
    label_counts = defaultdict(int)

    # Iterate through all .txt files in the directory
    for filename in os.listdir(directory):
        if filename.endswith(".txt"):
            filepath = os.path.join(directory, filename)
            with open(filepath, 'r') as file:
                lines = file.readlines()
                for line in lines:
                    label = line.split()[0]  # Get the label (first element in each line)
                    label_counts[label] += 1

    return label_counts

def summarize_counts(label_counts):
    summary = []
    for label, count in sorted(label_counts.items()):
        summary.append(f"{label} : {count} instance{'s' if count > 1 else ''}")
    return "\n".join(summary)
```

```python
# Directory containing the YOLO .txt files
directory = "/content/drive/MyDrive/YangBOT/Dataset/DatasetV2_67/train_test_val_original/train/labels"

# Count and summarize the labels
label_counts = count_yolo_labels(directory)
summary = summarize_counts(label_counts)

# Print the summary
print(summary)
```

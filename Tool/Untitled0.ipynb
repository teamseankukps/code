```python
import os

def count_files_in_folder(folder_path):
    # List all items in the folder
    items = os.listdir(folder_path)

    # Filter out only files (not directories)
    file_count = sum(1 for item in items if os.path.isfile(os.path.join(folder_path, item)))

    return file_count

# Example usage
folder_path = '/content/drive/MyDrive/YangBOT/Dataset/Dataset of Target Non-Target Problem/target'
print(f"Number of files in the folder: {count_files_in_folder(folder_path)}")
```

```python
folder_path2 = '/content/drive/MyDrive/YangBOT/Dataset/DatasetV2_67/train_test_val_original2/train/labels'
print(f"Number of files in the folder: {count_files_in_folder(folder_path2)}")
```
